1. Como é a distribuição do uso de força dentre as delegacias e os setores? Em cada setor, qual o *beat* com
maior número de incidentes? Apresente também o ranking dos setores segundo o percentual de incidentes

```python
# PRECINCT = DELEGACIAS OU DISTRITOS
# BEAT     = 3 SEÇÕES MENORES DE CADA SETOR
# SECTOR   = 17 SETORES NO TOTAL DA CIDADE
```

```python
DW[['BEAT_NUM','BEAT','PRECINCT_NUM','PRECINCT','INCIDENT_TYPE_LEVEL','INCIDENT_TYPE','SECTOR']].head(2)
```

```python

#Em cada setor, qual o beat com maior número de incidentes? 
DW = DW.sort_values(['PRECINCT'],ascending=False)
gp = DW.groupby(['SECTOR','BEAT'])
gp.size().get('B').max()

LISTA_SETORES = list(DW.SECTOR.drop_duplicates())
LISTA_BEATS = []
LISTA_VALORES = [] 
DIC_RESULT = {}

def plot_beat_com_maior_incidentes_de_cada_setor(grupo):
    for SETOR in LISTA_SETORES:
        LISTA_BEATS.append(gp.size().get(SETOR).sort_values(ascending=False)[0:1].keys()[0])
        LISTA_VALORES.append(gp.size().get(SETOR).max())
    DIC_RESULT = {'SECTOR': LISTA_SETORES, 'BEAT': LISTA_BEATS, 'INCIDENT':LISTA_VALORES}
    plt.figure(figsize=(20,7))
    sns.barplot(x="BEAT", y="INCIDENT",hue='SECTOR', data=pd.DataFrame(DIC_RESULT).sort_values(['INCIDENT']),palette='rainbow').set_title("BEATS COM MAIOR INCIDENTES DE CADA SETOR")
    
plot_beat_com_maior_incidentes_de_cada_setor(gp)
```

```python
# NUMERO DE INCIDENTES POR SEMANA EM CADA ANO


DW = DW.sort_values(['YEAR','WEEK_NUM'],ascending=True)
gp2 = DW.groupby(['YEAR','WEEK_NUM'])
anos = list(DW.YEAR.drop_duplicates())

cores =  ['blue','red','orange','green','yellow']

def criar_grafico_semana_por_anos(groupby,anos):
    cor = 0
    for ano in anos:
        grupo = groupby.size().get(ano)
        plt.plot( list(grupo.keys()), list(grupo.get_values()), 'go') 
        plt.plot( list(grupo.keys()), list(grupo.get_values()), 'k--',color=cores[cor],linewidth=2.0,label='ANO %d'%(ano)) 
        plt.legend()
        cor = cor+1
        
        
plt.figure(figsize=(18,8))
criar_grafico_semana_por_anos(gp2,anos)
plt.title("QTD DE INCIDENTES VS SEMANA DO ANO",fontsize=18)

plt.xlabel("NÚMERO DA SEMANA",fontsize=15)
plt.ylabel("QTD DE OCORRÊNCIA",fontsize=15)
```

```python
# NUMERO DE INCIDENTES POR MÊS EM CADA ANO


DW = DW.sort_values(['YEAR','MONTH'],ascending=True)
gp2 = DW.groupby(['YEAR','MONTH'])
anos = list(DW.YEAR.drop_duplicates())

cores =  ['blue','red','orange','green','yellow']

def criar_grafico_semana_por_anos(groupby,anos):
    cor = 0
    for ano in anos:
        grupo = groupby.size().get(ano)
        plt.plot( list(grupo.keys()), list(grupo.get_values()), 'go') 
        plt.plot( list(grupo.keys()), list(grupo.get_values()), 'k--',color=cores[cor],linewidth=2.0,label='ANO %d'%(ano)) 
        plt.legend()
        cor = cor+1
        
plt.figure(figsize=(18,8))
criar_grafico_semana_por_anos(gp2,anos)
plt.title("QTD DE INCIDENTES VS MÊS DO ANO",fontsize=18)

plt.xlabel("NÚMERO DO MES",fontsize=15)
plt.ylabel("QTD DE OCORRÊNCIA",fontsize=15)
```

```python
gg = gp_officer = DW.groupby(['SECTOR','OFFICER_ID']).size()
LISTA_SETORES = list(DW.SECTOR.drop_duplicates())
LISTA_PERCENTS = []
DICT_PERCENT = {}
def plot_percent_officer_reincident(group):
    for SETOR in LISTA_SETORES:
        percent = (100*group[group>1].get(SETOR).sum()) / group.get(SETOR).sum()
        LISTA_PERCENTS.append(round(percent,1))
        #print(' p : ',round(percent,1) , ' vv ', gg.size().get(SETOR)[2], ' t : ',sum(list(gg.size().get(SETOR))))
    DICT_PERCENT = {'SECTOR':LISTA_SETORES,'PERCENT_LEVEL_2':LISTA_PERCENTS}
    sns.barplot(x="SECTOR", y="PERCENT_LEVEL_2", data=DICT_PERCENT,palette='rainbow')
    
plt.figure(figsize=(18,5))
plot_percent_officer_reincident(gg)
```
